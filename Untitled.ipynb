```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
link='https://raw.githubusercontent.com/AshishJangra27/Machine-Learning-with-Python-GFG/main/Feature%20Engineering/Feature%20Selection/global_heat_index.csv'
```

```python
data=pd.read_csv(link)
```

```python
data.head()
```

```python
data.info()
```

```python
clean1=data[data['Hour']<6] 
clean2=data[data['Hour']>18]
clean1['Solar Radiation (GHI)'].unique()
clean2['Solar Radiation (GHI)'].unique()
```

```python
data=data.drop(clean1.index,axis=0)
data=data.drop(clean2.index,axis=0)
```

```python
data.shape
```

```python
x=data.iloc[:,0:10]
y=data.iloc[:,-1]
from sklearn.preprocessing import MinMaxScaler
x=pd.DataFrame(MinMaxScaler().fit_transform(x))
```

```python
x.head()
```

### Correlation Matrix with HeatMap

```python
corr_matrix=data.corr()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(data.corr(),annot=True)
```

### ExtraTreeClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(x,y)
```

```python
col=['Year','Month','Day','Hour','DewPoint','Temperature','Pressure','Relative Humidity','Wind Direction','Wind Speed']
impfea=pd.Series(model.feature_importances_,index=col)
impfea
```

```python
impfea.nlargest(10).plot(kind='bar')
plt.show()
```

### SelectKBest Method

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
Best_fea=SelectKBest(score_func=chi2,k=10).fit(x,y)
```

```python
scores=pd.Series(Best_fea.scores_,index=col)
scores
```

```python
scores.nlargest(10).plot(kind='bar')
plt.show()
```
